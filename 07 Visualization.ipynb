```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
#Step 3. Assign it to a variable called chipo.

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
chipo = pd.read_csv(url, sep = '\t')
```

```python
#Step 4. See the first 10 entries

chipo.head(10)
```

```python
#Step 5. Create a histogram of the top 5 items bought


item_counts = chipo['item_name'].value_counts()

top_items = item_counts.head(5)

plt.bar(top_items.index, top_items.values)

plt.xlabel('Items')

plt.ylabel('Purchase Count')

plt.title('Top 5 Items Bought at Chipotle')

plt.xticks(rotation=45)

plt.show()
```

```python
#Step 6. Create a scatterplot with the number of items orderered per order price


chipo.item_price = [float(value[1:-1]) for value in chipo.item_price]

# then groupby the orders and sum
orders = chipo.groupby('order_id').sum(numeric_only=True) 

plt.scatter(x=orders.quantity, y=orders.item_price, s=50, c='green')

# Set the title and labels

plt.xlabel('Items ordered')
plt.ylabel('Price')
plt.title('items ordered per order price')

# Set the y-axis minimum to 0
plt.ylim(0)
plt.show()
```

```python
#BONUS: Create a question and a graph to answer your own question.

#What are the top 10 most popular items ordered in the dataset, and what is the distribution of their quantities?




item_counts = chipo['item_name'].value_counts()


top_10_items = item_counts.head(10)

plt.figure(figsize=(10, 6))

plt.bar(top_10_items.index, top_10_items.values, color='skyblue')

plt.xticks(rotation=45, ha='right')

plt.xlabel('Item Name')

plt.ylabel('Quantity Ordered')

plt.title('Top 10 Most Popular Items and Their Quantities')

plt.tight_layout()

plt.show()


```
